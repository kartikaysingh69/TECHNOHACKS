```python
import pandas as pd
import matplotlib.pyplot as plt

# File path for the Iris's dataset:
iris_file_path = r'C:\Users\karti\OneDrive\Desktop\Iris.csv'
# Iris's dataset:
iris_data = pd.read_csv(iris_file_path)
# Column name:
column_name = 'SepalLengthCm'
plt.figure(figsize=(4, 4))
plt.hist(iris_data[column_name], bins=100, color='red', edgecolor='blue')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title(f'Histogram of {column_name}')
plt.show()
```
